```python
correlation_matrix = df[['missions', 'days in space', 'spacewalks', 'days spacewalking']].corr()

# Korelasyon matrisini görselleştirme
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasyon Matrisi')
plt.show()
```

```python
# Şirketlere göre ortalama görev süreleri
company_avg_days = df.groupby('company space agency')['days in space'].mean().sort_values(ascending=False)

# Görselleştirme
plt.figure(figsize=(12, 6))
sns.barplot(x=company_avg_days.values, y=company_avg_days.index, palette='viridis')
plt.xlabel('Ortalama Görev Süresi (days)')
plt.title('Şirketlere Göre Ortalama Görev Süreleri')
plt.show()
```

```python
# Ülkelerin toplam görev sürelerini hesapla
country_total_days = df.groupby('country')['days in space'].sum().sort_values(ascending=False)

# Grafik çizimi
plt.figure(figsize=(12, 6))
sns.barplot(x=country_total_days.values, y=country_total_days.index, palette='viridis')
plt.title('Ülkelere Göre Toplam Uzayda Geçirilen Gün Sayısı')
plt.xlabel('Toplam Gün Sayısı')
plt.ylabel('Ülke')
plt.show()
```
